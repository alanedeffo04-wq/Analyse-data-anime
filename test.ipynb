```python
# 1. Créer une copie pour travailler proprement
df_clean = df.copy()

# 2. Supprimer les lignes où il manque des infos cruciales (ex: le nom de l'anime)
df_clean = df_clean.dropna(subset=['Anime'])

# 3. Supprimer les doublons
df_clean = df_clean.drop_duplicates()

# 4. Afficher le résultat
print("Nettoyage terminé ! ✅")
print(f"Lignes restantes : {len(df_clean)}")
display(df_clean.head())
```

ENRICHIR LES DONNEES

```python
import datetime
current_year = datetime.datetime.now().year

# On extrait l'année de la colonne Date_Pub et on calcule l'âge
df_clean['Annee_Sortie'] = pd.to_datetime(df_clean['Date_Pub']).dt.year
df_clean['Age_Anime'] = current_year - df_clean['Annee_Sortie']
display(df_clean.head(61))

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Calcul du Score de Synthèse (Qualité Pure)
# Note Globale (50%) + Régularité (30%) + Meilleur Épisode (20%)
df_clean['Score_Synthese'] = (
    (df_clean['Note_Globale'] * 0.5) +
    (df_clean['Regularite'] * 0.3) +
    (df_clean['Note_Meilleur_Ep'] * 0.2)
)

# 2. Tri pour obtenir les 10 meilleurs
top_10_final = df_clean.sort_values(by='Score_Synthese', ascending=False).head(10)

# 3. Visualisation
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

# On utilise une palette élégante 'crest'
ax = sns.barplot(
    x='Score_Synthese',
    y='Anime',
    data=top_10_final,
    palette='crest',
    hue='Anime',
    legend=False
)

# 4. Personnalisation
plt.title('TOP 10 GLOBAL : Qualité, Pics et Constance', fontsize=16, fontweight='bold')
plt.xlabel('Score de Synthèse (Moyenne Pondérée / 10)', fontsize=12)
plt.ylabel('Titre de l\'Anime', fontsize=12)

# Zoom sur l'axe X pour voir les différences (souvent dans les 8-9)
plt.xlim(top_10_final['Score_Synthese'].min() - 0.2, 10)

# 5. Ajout des notes au bout des barres
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=5)

plt.tight_layout()
plt.show()

# 6. Affichage du classement textuel
print("🏆 LE CLASSEMENT ULTIME 🏆")
print(top_10_final[['Anime', 'Score_Synthese', 'Profil']].to_string(index=False))
```
